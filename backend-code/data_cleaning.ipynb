```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Setting up the DataFrame for salaries
salaries_csv = Path('Resources/nba_salaries/nba_2022-23_all_stats_with_salary.csv')
salaries_df = pd.read_csv(salaries_csv)
salaries_df.head(20)
```

```python
salaries_df.info()
```

```python
clean_salaries_df = salaries_df.drop(columns=['MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FT%','ORB','DRB','Total Minutes','3PAr','FTr','ORB%','DRB%','TRB%','AST%','STL%','BLK%','TOV%'])
clean_salaries_df = clean_salaries_df.dropna(how='any')
clean_salaries_df.info()
```

```python
# Remove the / from players that were traded mid season and count the last team they were on
for index, row in clean_salaries_df.iterrows():
    # Check if '/' exists in the 'Team' column of the current row
    if '/' in row['Team']:
        # Remove the first 4 characters from the 'Team' column of the current row
        clean_salaries_df.at[index, 'Team'] = row['Team'][4:]

# Display the updated DataFrame
print(clean_salaries_df.head())
```

```python
# Create a new column to decide if a player was a starter or not
for index, row in clean_salaries_df.iterrows():
    # If they started half of their games, they are listed as a starter
    started_ratio = row['GS'] / row['GP']
    if started_ratio >= .5:
        starter = True
    else:
        starter = False
    # Creates a new column for the starters
    clean_salaries_df.at[index, 'Starter?'] = starter

clean_salaries_df.head()
```

```python
# Setting up a grouped dataframe organized by team and position
grouped_df = clean_salaries_df.sort_values(['Team']).reset_index(drop=True).groupby(['Team', 'Position'])
grouped_df.head()
```
